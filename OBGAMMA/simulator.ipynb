```python
def run_with_time(func, *args):

    def wrapper(*args):
        import time
        st_time = time.time()
        return_value = func(*args)
        ed_time = time.time()
        elapsed_time = ed_time - st_time
        elapsed_hours = int(elapsed_time // 3600)
        elapsed_mins = int((elapsed_time - elapsed_hours * 3600) // 60)
        elapsed_secs = elapsed_time - elapsed_hours * 3600 - elapsed_mins * 60
        print(f"Elapsed Time: {elapsed_hours} hours, {elapsed_mins} mins, {elapsed_secs:.2f} seconds")
        return return_value

    return wrapper

@run_with_time
def test(n):
    return [100**100 for i in range(1000) for j in range(1000)]


x = test(100)
```
